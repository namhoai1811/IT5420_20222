```python
compare_cl = recordlinkage.Compare()
compare_cl.string("title", "title", method="levenshtein", threshold=0.85, label="title")
compare_cl.string("description", "description", method="levenshtein", threshold=0.7, label="title")
# compare_cl.string("direct", "direct", method="levenshtein", threshold=0.9, label="title")
compare_cl.exact("province", "province", label="province")
# compare_cl.exact("street", "street", label="street")
compare_cl.exact("floor", "floor", label="floor")

features = compare_cl.compute(candidates, dataset)
```
